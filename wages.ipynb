```python
import pandas as pd
```

```python
wages = pd.read_csv('wages.csv')
```

```python
wages.head()
```

Data Cleaning & Transformation

Check for Null Values

```python
wages.isnull().sum()
```

Replace Null Values

```python
wages['City'].fillna('Ottawa', inplace = True)
```

```python
wages['Job Title'].fillna('Data Analyst', inplace = True)
```

```python
fill_median = wages['Latitude'].median()
wages['Latitude'].fillna(fill_median, inplace = True)
```

```python
wages.head()
```

```python
wages.isnull().sum()
```

Data Sync 

Replacing the gender 'Male / Female' with M or F

```python
wages['gender'] = wages['gender'].replace(['Male', 'Female'], ['M', 'F'])
```

Adding 'full_name' column 

```python
wages['full_name'] = wages['first_name'] + ' ' + wages['last_name']
```

```python
wages.head()
```

Check for Duplicates

```python
wages[wages.duplicated('last_name')]
```

Remove duplicates by last name

```python
wages.drop_duplicates('last_name', keep = 'last')
```

Data Analysis

```python
# Top 10 Jobs sorted by Salary

wages.sort_values('Salary', ascending=False) .head(10)
```

```python
# Employee with the Highest Salary

wages[wages['Salary'] == wages['Salary'].max()]
```

```python
# Employee with the Lowest Salary

wages[wages['Salary'] == wages['Salary'].min()]
```

```python
# Top 10 Job Titles by no. of employees

wages['Job Title'].value_counts().head(10)
```

```python
# Create a function to remove '$' from the salary and then cast it to float

def clean_salary(data):
    salary = data['Salary'].replace('$',"")
    salary = float(salary)
    return salary
```

```python
# Add a new salary column for further analysis after applying the 'clean_salary' function

wages['salary_new'] = wages.apply(clean_salary, axis = 1)
```

```python
wages.info()
```

```python
# Top 10 Job Titles by Average Salary

wages.groupby('Job Title')['salary_new'].mean().sort_values(ascending = False).head(10)
```

```python
# Male vs Female Average Salary
 
wages.groupby('gender')['salary_new'].mean()
```

```python
# Split male and female salaries into new separate columns for further analysis

def male_salary(data):
    if data['gender'] == 'Male':
        return data ['salary_new']
wages['male_salary'] = wages.apply(male_salary, axis = 1)

def female_salary(data):
    if data['gender'] == 'Female':
        return data ['salary_new']

wages['female_salary'] = wages.apply(female_salary, axis = 1)
```

```python
# Male vs Female Average Salary by Job Title

wages.groupby('Job Title')['salary_new', 'male_salary', 'female_salary'].mean()
```

```python
# Male vs Female Average Salary by Job Title and City

wages.groupby(['Job Title', 'City'])['salary_new', 'male_salary', 'female_salary'].mean()
```

```python
# No. of Data jobs

sum(wages['Job Title'].str.contains('Data'))
```

```python
# List of Data jobs sorted by Salary

wages[wages['Job Title'].str.contains('Data')].sort_values('Salary', ascending=False)
```
